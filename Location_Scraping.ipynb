```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
data = requests.get("https://unstats.un.org/sdgs/indicators/regional-groups/")
if data.status_code == 200:
    soup = BeautifulSoup(data.content, "html.parser")
    root_ul = soup.find('ul', id='tree')
```

```python
nums = [0, 0, 0, 0, 0, 0]
output = []
output1 = []
def extract_location_with_structure(element, indent=0):
    if element.name == 'h5':
        check = indent//2
        if check == 1:
            nums[0] += 1
            for i in range(1, len(nums)):
                nums[i] = 0
        elif check == 2:
            nums[1] += 1
            for i in range(2, len(nums)):
                nums[i] = 0
        elif check == 3:
            nums[2] += 1
            for i in range(3, len(nums)):
                nums[i] = 0
        elif check == 4:
            nums[3] += 1
            for i in range(4, len(nums)):
                nums[i] = 0
        elif check == 5:
            nums[4] += 1
            for i in range(5, len(nums)):
                nums[i] = 0
        elif check == 6:
            nums[5] += 1
            for i in range(6, len(nums)):
                nums[i] = 0
        numbering = '.'.join(str(nums[j]) for j in range(check))
        text = element.get_text(strip=False)

        parts = re.split(r'[\s]{5,}|\n+', text, maxsplit=1)

        if len(parts) == 2:
            location, code = parts
            location = location.strip()
            code = code.strip()
        else:
            location = text.strip()
            code = ''
        print(' ' * indent, numbering, location, code)
        output.append((check, numbering, location, code))
        output1.append([numbering, location])


    for child in element.find_all(recursive=False):
        extract_location_with_structure(child, indent + 1)

extract_location_with_structure(root_ul)
```

```python
df = pd.DataFrame(output, columns=["Indentation Level", "Structure Level", "Location", "Location Code"])

df.to_excel('Location_Extraction.xlsx', index=False)
```
